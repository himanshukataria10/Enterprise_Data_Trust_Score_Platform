```python
import pandas as pd
import numpy as np

print("Libraries Imported Successfully")
```

```python
dirty_df = pd.read_csv("../data/processed/dirty_employee_data.csv")

dirty_df.head()
```

```python
dirty_df.shape
```

```python
dirty_df.info()
```

```python
dirty_df.isnull().sum()
```

```python
dirty_df["Employee_ID"].duplicated().sum()
```

```python
dirty_df[
    (dirty_df["Salary"] < 25000) |
    (dirty_df["Salary"] > 120000)
]
```

```python
valid_departments = [
    "IT",
    "HR",
    "Finance",
    "Sales",
    "Marketing",
    "Operations"
]

dirty_df[
    ~dirty_df["Department"].isin(valid_departments)
]
```

```python
dirty_df["Joining_Date"] = pd.to_datetime(dirty_df["Joining_Date"])

dirty_df[
    dirty_df["Joining_Date"] > pd.Timestamp.today()
]
```

```python
clean_df = dirty_df.copy()

clean_df.head()
```

```python
clean_df[clean_df["Email"].isnull()]
```

```python
clean_df["Email"] = clean_df["Email"].fillna("unknown@company.com")
```

```python
clean_df["Email"].isnull().sum()
```

```python
clean_df[
    clean_df["Employee_ID"].duplicated(keep=False)
]
```

```python
clean_df = clean_df.drop_duplicates(
    subset="Employee_ID",
    keep="first"
)
```

```python
clean_df["Employee_ID"].duplicated().sum()
```

```python
clean_df[
    (clean_df["Salary"] < 25000) |
    (clean_df["Salary"] > 120000)
]
```

```python
clean_df.loc[
    clean_df["Salary"] < 25000,
    "Salary"
] = 25000
```

```python
clean_df[
    (clean_df["Salary"] < 25000) |
    (clean_df["Salary"] > 120000)
]
```

```python
clean_df[
    (clean_df["Salary"] < 25000) |
    (clean_df["Salary"] > 120000)
]
```

```python
clean_df.loc[
    ~clean_df["Department"].isin(valid_departments),
    "Department"
] = "Operations"
```

```python
clean_df[
    ~clean_df["Department"].isin(valid_departments)
]
```

```python
clean_df[
    clean_df["Joining_Date"] > pd.Timestamp.today()
]
```

```python
clean_df.loc[
    clean_df["Joining_Date"] > pd.Timestamp.today(),
    "Joining_Date"
] = pd.Timestamp.today().normalize()
```

```python
clean_df[
    clean_df["Joining_Date"] > pd.Timestamp.today()
]
```

```python
print("Missing Emails:", clean_df["Email"].isnull().sum())

print("Duplicate Employee IDs:",
      clean_df["Employee_ID"].duplicated().sum())

print("Invalid Salary:",
      len(
          clean_df[
              (clean_df["Salary"] < 25000) |
              (clean_df["Salary"] > 120000)
          ]
      ))

print("Invalid Department:",
      len(
          clean_df[
              ~clean_df["Department"].isin(valid_departments)
          ]
      ))

print("Future Joining Date:",
      len(
          clean_df[
              clean_df["Joining_Date"] > pd.Timestamp.today()
          ]
      ))
```

```python
clean_df.to_csv(
    "../data/processed/clean_employee_data.csv",
    index=False
)

print("Clean Dataset Saved Successfully")
```
