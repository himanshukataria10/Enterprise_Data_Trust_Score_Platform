```python
plt.figure(figsize=(10,5))

plt.hist(
    clean_df["Salary"],
    bins=10
)

plt.title("Employee Salary Distribution")

plt.xlabel("Salary")

plt.ylabel("Number of Employees")

plt.show()
```

```python
department_counts = clean_df["Department"].value_counts()

department_counts
```

```python
plt.figure(figsize=(8,5))

department_counts.plot(kind="bar")

plt.title("Employees by Department")

plt.xlabel("Department")

plt.ylabel("Number of Employees")

plt.show()
```

```python
city_counts = clean_df["City"].value_counts()

city_counts
```

```python
plt.figure(figsize=(8,5))

city_counts.plot(kind="bar")

plt.title("Employees by City")

plt.xlabel("City")

plt.ylabel("Number of Employees")

plt.show()
```

```python
clean_df["Trust_Category"].value_counts()
```

```python
trust_category = clean_df["Trust_Category"].value_counts()

plt.figure(figsize=(6,6))

trust_category.plot(kind="pie", autopct="%1.1f%%")

plt.ylabel("")

plt.title("Trust Category Distribution")

plt.show()
```
