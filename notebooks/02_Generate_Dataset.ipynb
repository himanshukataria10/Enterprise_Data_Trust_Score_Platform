```python
import random
from faker import Faker

fake = Faker("en_IN")
```

```python
departments = [
    "Human Resources",
    "Finance",
    "Sales",
    "Marketing",
    "IT",
    "Operations"
]

job_titles = [
    "HR Executive",
    "Finance Analyst",
    "Sales Executive",
    "Marketing Executive",
    "Software Engineer",
    "Data Analyst"
]

cities = [
    "Bengaluru",
    "Mumbai",
    "Delhi",
    "Hyderabad",
    "Pune",
    "Jaipur"
]
```

```python
employee_data = []
```

```python
for employee_id in range(1001,1051):

    employee = {

        "Employee_ID": employee_id,

        "First_Name": fake.first_name(),

        "Last_Name": fake.last_name(),

        "Email": fake.email(),

        "Phone": fake.phone_number(),

        "Department": random.choice(departments),

        "Job_Title": random.choice(job_titles),

        "Salary": random.randint(25000,120000),

        "Joining_Date": fake.date_between(
            start_date="-5y",
            end_date="today"
        ),

        "City": random.choice(cities)

    }

    employee_data.append(employee)
```

```python
employee_df = pd.DataFrame(employee_data)
```
