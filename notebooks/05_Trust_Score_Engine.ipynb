```python
clean_df["Trust_Score"] = 100
```

```python
def calculate_trust_score(row):

    score = 100

    # Missing Email
    if pd.isna(row["Email"]):
        score -= 20

    # Invalid Salary
    if row["Salary"] < 25000 or row["Salary"] > 120000:
        score -= 20

    # Invalid Department
    valid_departments = [
        "IT",
        "HR",
        "Finance",
        "Sales",
        "Marketing",
        "Operations"
    ]

    if row["Department"] not in valid_departments:
        score -= 15

    # Future Joining Date
    joining_date = pd.to_datetime(row["Joining_Date"])

    if joining_date > pd.Timestamp.today():
        score -= 15

    return max(score, 0)
```

```python
clean_df["Trust_Score"] = clean_df.apply(
    calculate_trust_score,
    axis=1
)
```

```python
clean_df.head()
```

```python
clean_df["Trust_Score"].describe()
```

```python
def trust_category(score):
    if score >= 90:
        return "Excellent"
    elif score >= 75:
        return "Good"
    elif score >= 50:
        return "Fair"
    else:
        return "Poor"

clean_df["Trust_Category"] = clean_df["Trust_Score"].apply(trust_category)
```

```python
clean_df[["Employee_ID", "Trust_Score", "Trust_Category"]].head()
```

```python
clean_df.to_csv(
    "../data/processed/clean_employee_data.csv",
    index=False
)
```
